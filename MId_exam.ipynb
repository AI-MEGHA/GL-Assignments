# STEP1- Importing necessary libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

# STEP 2- Data Preprocessing

```python
df=pd.read_csv("data1.csv")
```

```python
df.head()
```

```python
df.nunique()
```

# STEP 3- Data Cleaning

Removing $ and converting amount to integer type

```python
df['Amount']=df.Amount.apply(lambda x: x.replace('-',''))
df
```

```python
df['Amount']=df.Amount.apply(lambda x: x.replace('$',''))
df
```

```python
df['Amount']=df.Amount.apply(lambda x: x.replace(',',''))
df.head()
```

```python
df['Amount'] = df['Amount'].astype('float')
```

```python
#We are adding a sales column for furthur analysis of sales with other columns
#sales= Amount*Units
df['Sales']=df.Amount*df.Units
df.head()
```
